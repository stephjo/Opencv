<b>Reading image</b>

cv2.IMREAD_COLOR- To load a color image neglecting existing transparency (default flag)<br/>
cv2.IMREAD_GRAYSCALE- To load a grayscale image<br/>
cv2.IMREAD_UNCHANGED- To load an image including an alpha channel<br/>
We can pass integers 1, 0, or -1.

```python
img=cv2.imread("~/flower.jpg",0)
cv2.imshow('Flower',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Working in scripts, a call to <b>waitKey(0)</b> is beneficial. This is a keyboard-binding function/method with time in milliseconds. This function waits for certain milliseconds for a keyboard event, within which, if we press any key, the program continues. When we pass 0, we make it wait indefinitely for a keystroke. We can also make it wait for specific keys.

<b>cv2.destroyAllWindows()</b> is another function/method to destroy all windows we created. <b>cv2.destroyWindow()</b> destroys a specific window.

<b>Writing image</b>

```python
cv2.imwrite('Flower_gray.jpg',img)
```

We can display this image using <b>Matplotlib</b>.

```python
import matplotlib.pyplot as plt
#function to display image using matplotlib
def showimg(img):
    plt.imshow(img,cmap='gray',interpolation='bicubic')
    plt.xticks([]),plt.yticks([])
    plt.show()
showimg(img)    
```

<b>Drawing Lines in Python</b>

cv2.line(img, Point pt1, Point pt2, color[,thickness[,lineType[,shift]]]) > img<br/><br/>
PARAMETERS:	<br/>
img  Image.<br/>
pt1  First point of the line segment.<br/>
pt2  Second point of the line segment.<br/>
color  Line color.<br/>
thickness  Line thickness.<br/>
lineType <br/>
Type of the line:<br/>
LINE_8 (or omitted)  8-connected line.<br/>
LINE_4  4-connected line.<br/>
LINE_AA  antialiased line.<br/>
shift  Number of fractional bits in the point coordinates.

```python
import numpy as np
# Create a black image
img=np.zeros((512,512,3),np.uint8)
```

```python
cv2.line(img,(0,0),(511,511),(255,0,0),5)
cv2.imshow("img",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#print line
showimg(img)
```

<b>Drawing a Circle:</b>

cv2.circle(img, center, radius, color[, thickness[, lineType[, shift]]])  img<br/>
PARAMETERS:	<br/>
img  Image where the circle is drawn.<br/>
center  Center of the circle.<br/>
radius  Radius of the circle.<br/>
color  Circle color.<br/>
thickness  Thickness of the circle outline, if positive. Negative thickness means that a filled circle is to be drawn.<br/>
lineType  Type of the circle boundary.<br/>
shift  Number of fractional bits in the coordinates of the center and in the radius value.

```python

# Create a black image
img = np.zeros((512,512,3), np.uint8)

# Draw a red closed circle
cv2.circle(img,(200,200), 40, (0,0,255), 2)

cv2.imshow("img",img)

cv2.waitKey(0)
#print circle
showimg(img)
```

<b>Drawing an Ellipse:</b>

cv2.ellipse(img, center, axes, angle, startAngle, endAngle, color[, thickness[, lineType[, shift]]])  img <br/>
cv2.ellipse(img, box, color[, thickness[, lineType]])  img<br/>
PARAMETERS:	<br/>
img  Image.<br/>
center  Center of the ellipse.<br/>
axes  Half of the size of the ellipse main axes.<br/>
angle  Ellipse rotation angle in degrees.<br/>
startAngle  Starting angle of the elliptic arc in degrees.<br/>
endAngle  Ending angle of the elliptic arc in degrees.<br/>
box  Alternative ellipse representation via RotatedRect or CvBox2D. This means that the function draws an ellipse inscribed in the rotated rectangle.<br/>
color  Ellipse color.<br/>
thickness  Thickness of the ellipse arc outline, if positive. Otherwise, this indicates that a filled ellipse sector is to be drawn.<br/>
lineType  Type of the ellipse boundary.<br/>
shift  Number of fractional bits in the coordinates of the center and values of axes.<br/>

```python
# Create a black image
img = np.zeros((512,512,3), np.uint8)

# Draw ellipse
cv2.ellipse(img,(200,200),(80,100),0,0,360,(0,255,0),1)
cv2.ellipse(img,(200,200),(80,50),45,0,360,(255,0,0),-1)

#Display the image
cv2.imshow("img",img)

cv2.waitKey(0)

#print ellipse
showimg(img)
```

<b>Drawing a Rectangle:</b><br/><br/>
cv2.rectangle(img, pt1, pt2, color[, thickness[, lineType[, shift]]])  img<br/>
PARAMETERS:	<br/>
img  Image.<br/>
pt1  Vertex of the rectangle.<br/>
pt2  Vertex of the rectangle opposite to pt1 .<br/>
rec  Alternative specification of the drawn rectangle.<br/>
color  Rectangle color or brightness (grayscale image).<br/>
thickness  Thickness of lines that make up the rectangle. Negative values, like CV_FILLED , mean that the function has to draw a filled rectangle.<br/>
lineType  Type of the line.<br/>
shift  Number of fractional bits in the point coordinates.

```python

# Create a black image
img = np.zeros((512,512,3), np.uint8)

# Draw rectangle(not filled)
cv2.rectangle(img,(15,20),(70,50),(0,255,0),3)

# Draw filled rectangle
cv2.rectangle(img,(256,256),(170,150),(255,0,0),-1)

#Display the image
cv2.imshow("img",img)

cv2.waitKey(0)

#print rectangle
showimg(img)
```

<b>Drawing a Polygon:</b><br/>
cv2.polylines(img, pts, isClosed, color[, thickness[, lineType[, shift]]])  img<br/>
PARAMETERS:	<br/>
img  Image.<br/>
pts  Array of polygonal curves.<br/>
npts  Array of polygon vertex counters.<br/>
ncontours  Number of curves.<br/>
isClosed  Flag indicating whether the drawn polylines are closed or not. If they are closed, the function draws a line from the last vertex of each curve to its first vertex.<br/>
color  Polyline color.<br/>
thickness  Thickness of the polyline edges.<br/>
lineType  Type of the line segments.<br/>
shift  Number of fractional bits in the vertex coordinates.    

```python
# Create a black image
img = np.zeros((512,512,3), np.uint8)

# Draw a polygon
pts = np.array([[10,5],[20,30],[100,400],[300,300]], np.int32)
pts = pts.reshape((-1,1,2))
cv2.polylines(img,[pts],True,(0,255,255))

#Display the image
cv2.imshow("img",img)

cv2.waitKey(0)

#print polygon
showimg(img)
```

<b>Putting Text in Images</b>

```python
font=cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,'OpenCV',(10,500),font,4,(255,255,23),2,cv2.LINE_AA)
cv2.imshow('image',img)

cv2.waitKey(0)
cv2.destroyAllWindows()
#print image
showimg(img)
```
